```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='label')
plt.xticks(rotation=90)
plt.title('Distribution of Labels')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='kids')
plt.title('Distribution of Kids Flag')
plt.show()
```

```python
print(f"Number of unique sender IDs: {df['sender_id'].nunique()}")
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['sender_id'], kde=True, bins=30)
plt.title('Distribution of Sender IDs')
plt.show()
```

```python
print(f"Number of rows before removing 'Not sure' labels: {len(df)}")
```
